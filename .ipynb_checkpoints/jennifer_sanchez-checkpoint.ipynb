```python
import matplotlib.pyplot as plt # importa la biblioteca matplotlib pyplot y se le asigna un alias "plt" para acceder a su funcionabilidad
import matplotlib.patches as mpatches# importa el módulo mpatches de la biblioteca matplotlib
import numpy as np# importa numpy con el alias np
d={}#crea un diccionario vacío
# con el ciclo for siguiente se recorre los datos del diccionario data en cerveza,y añade las llaves que son los diferentes paises y una lista de sus correspondientes precios
for br in data["cerveza"]:
      d.setdefault(br["marca"],[br["precio"]]),
      if br["marca"] in d: 
        d[br["marca"]].append(br["precio"])         
#crea dos listas vacías 
keys=[] 
values=[]
# mediante el siguiente ciclo for se añade a la lista keys las diferentes marcas en el ddiccionario y a la lista values los valores de las anteriores marcas
for key in d:
        keys.append(key)
        values.append(d[key])  
# mediante el siguiente ciclo for añade a la lista brand las marcas de cerveza que estan en el diccionario data
brand=[]
price=[]
for marc in data["cerveza"]:
       brand.append(marc["marca"])
for prec in data["cerveza"]: # se asigna prec a cada elemento  en el diccionario  data en cerveza, mediante un ciclo for
       price.append(prec["precio"])# por cada prec, se le añade a la lista el precio de cada cerveza en el diccionario data        
# igualo las variables x y y a las listas brand y price respectivamente
x=brand
y=price
# dibuja un diagrama de disperción con los datos x y y, donde los colores de los puntos se basan en y y se asignan utilizando el mapa de colores "Spectral"
plt.scatter(y,x, c=y,cmap="Spectral")
plt.show()# muestra el gráfico de disperción 
# crea la función half que calcula el promedio        
def half(l,h=0):
    for i in l:
        h+=i
    c=int(h/len(l))
    return c 
# crea la función max_min que calcula el máximo y el mínimo
def max_min(l):
   l.sort()
   return [l[0],l[-1]] 
#crea una lista vacía en la variable v 
v=[]
# añade a la lista v los promedios del precio de las marcas
for i in values:
    v.append(half(i))
#clono la lista v en z
z=v[:]    
# les da valores a las variables a y b de las posiciónes de la función max_min en 0 y 1 en z respectivamente
a=z.index(max_min(v)[-1])
b=z.index(max_min(v)[0])
# me imprime cual es la marca que mayor promedio que tiene, la que menos , y cuales son
print(keys[a],"con",max_min(v)[-1],"es la marca con mayor promedio ",keys[b],"con",max_min(v)[0]," es la marca con menor promedio de precio")    
```

### Los refrescos de botellas de plástico:
-poseen acetaldehído que podrían pasar al refresco y afectar su sabor, aún en cantidades mínimas.

-el plástico es más permeable al CO2, la efervecencia de la bebida se escapa más rápido.

-no se puede reciclar en más botellas (si a caso en fibras de alfombras, ropas o sacos de dormir).

-las personas no reciclan botellas de plástico tan a menudo.

### Los refrescos de lata:

-son menos permeables al CO2, conservan las burbujas por más tiempo .

-proviene de la roca bausita, y su extracción puede arruinar ecosistemas.

-se pueden reciclar en nuevas latas.

-se estima que casi el 75% de aluminio producido está en uso.

Alguna vez han sentido que el día es muy caluroso, y van a una cafetería y da la casualidad que no hay cerveza, pues sabían que según los resultados de mi estudio aproximadamente el 8% de las cafeterías que tienen refresco no tienen cerveza, lo cual vemos reflejado en el mapa siguiente, de color rojo son los puntos que tienen refresco, y azul cerveza.

```python
import folium # importa folium
from folium.plugins import MiniMap# importa el módulo MiniMap de la biblioteca folium.plugins
import json # importa la  biblioteca json para manipular datos json
f=open("proyecto_final_cd.json")# abre el archivo "proyecto_final_cd.json" en modo lectura y le asigna la variable "f"
data=json.load(f) # carga el contenido de "f" en la variable "data" utilizando la funcion json.load(f)
mp=folium.Map(location=[ 23.16592025,-82.28157043],zoom_star=3)# define un punto de eje de coordenadas, con la funcion location y un zoom_star para la cercanía
#crea dos listas vacías y una variable igual a cero
lc=[]
lo=[]
c=0
# recorre el rango del tamaño de refresco en el diccionario data, y recorre rango del tamaño de cerveza, y busca cuando no exista ningún púnto de cerveza con la misma localización del refresco, y si lo encuentra le suma 1 a la variable c
for j in range(len(data["refresco"])): 
    for i in range(len(data["cerveza"])): 
       if data["cerveza"][i]["imagenes"]==data["refresco"][j]["imagenes"]:
            break
    else:
       c+=1
# imprime la cantidad de lugares con refresco sin cerveza y su por ciento que representan del total de puntos en común 
print(c,"puestos en los cuales no hay cerveza",int((c*100)/(len(data["cerveza"]))),"por ciento del total de puestos de venta de refrescos")        
#mediante los siguientes ciclos for  se añade la localización a las listas de refresco y cerveza respectivamente
for loc in data["refresco"]:
       lo.append(loc["loc"])
for loc in data["cerveza"]:
       lc.append(loc["loc"])
# mediante el siguiente ciclo for se añaden las localizaciones a marcadores, y en el caso de el for en lc, se le agrega al marcador que sea de color rojo        
for i in lo:
    folium.Marker(location=i).add_to(mp)
for i in lc:
    folium.Marker(location=i,icon=folium.Icon(color="red")).add_to(mp)
mp    #carga el mapa
```
