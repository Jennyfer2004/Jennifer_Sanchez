### Los refrescos de plástico:
-poséen acetaldehído que podrían pasar al refresco y afectar su sabor, aún en cantidades mínimas.

-el plástico es más permeable al CO2, la efervecencia de la bebida se escapa más rápido.

-se derivan de productos no renovables(petróleo).

-no se puede reciclar en más botellas (si a caso en fibras de alfombras, ropas o sacos de dormir).

-las personas no reciclan botellas de plástico tan a menudo.

### Los refrescos de lata:
-el interior de las latas está revestido con un polímero que puede contener BPA o Bishfenol A, con lo cual se evita el contacto del refresco con el metral.

-son menos permeables al CO2, conserban las burbujas por más tiempo .

-proviene de la roca bausita, y su extracción puede arruinar ecosistemas.

-se pueden reciclar en nuevas latas.

-se estima que casi el 75% de aluminio producido está en uso.
